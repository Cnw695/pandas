```python
import pandas as pd
import numpy as np

file = 'students_complete.csv'
students = pd.read_csv(file)
print(students.head())

file2 = 'schools_complete.csv'
schools = pd.read_csv(file2)
print(schools.head())
```

```python
df = pd.merge(students, schools, on='school_name')
#df[df.type == "District"]
df
```

```python
totalschools = len(df['school_name'].unique())
totalschools
```

```python
totalstudent = df['Student ID'].count()
totalstudent
```

```python
totalbudget = df['budget'].unique().sum()
totalbudget
```

```python
averagemath = df['math_score'].mean()
averagemath
```

```python
averageread = df['reading_score'].mean()
averageread
```

```python
#df.dtypes

```

```python
#this shouldnt be more than 26976. that is the number of District Students
#district_pass = df['math_score'].mean()
#district_pass

math = df[df['math_score']>=70].count()["student_name"] /  totalstudent
math
#['gender','math_score']]

```

```python
reading = df[df['reading_score']>=70].count()["student_name"] /  totalstudent
reading
```

```python
overall = (math + reading) / 2
overall
```

```python
district_summary = pd.DataFrame([{'Total Schools': totalschools, 'Total Students': totalstudent, 'Total Budget': totalbudget,
                       'Average Math': averagemath, 'Average Read': averageread, 'Math Pass %': math, 'Reading Pass %': reading, 'Overall Pass %': overall}
                     ])
district_summary
```

# school_summary

```python
school_names = schools.set_index(['school_name'])['type']
school_names
```

```python
df['math_passing'] = df.loc[df['math_score']>=70, 'math_score']
df['reading_passing'] = df.loc[df['reading_score']>=70, "reading_score"]
df_grouped = df.groupby(['school_name'])
student_count = df_grouped['Student ID'].count()
df
```

```python
budget = df_grouped['budget'].mean()
budget.map('${:,.2f}'.format)
```

```python
budget_perstudent = (budget / student_count).map('${:.2f}'.format)
budget_perstudent
```

```python
avgmath = df_grouped['math_score'].mean().map('{:.2f}%'.format)
avgreading = df_grouped['reading_score'].mean().map('{:.2f}%'.format)
avgmath
```

```python
percent_passmath = df_grouped['math_passing'].count() / student_count * 100
percent_passmath.map('{:.2f}%'.format)
```

```python
percent_passreading = df_grouped['reading_passing'].count() / student_count * 100
percent_passreading.map('{:.2f}%'.format)
```

```python
overallschool = (percent_passreading + percent_passmath) / 2
overallschool.map('{:.2f}%'.format)
```

```python
school_summary = pd.DataFrame({'School Type': school_names, 'Total Students': student_count.map('{:,}'.format),
                               'Total School Budget': budget.map('${:,.2f}'.format), 'Per Student Budget': budget_perstudent,
                              'Average Math Score': avgmath, 'Average Reading Score': avgreading,
                              '% Passing Math': percent_passmath.map('{:.2f}%'.format), '% Passing Reading': percent_passreading.map('{:.2f}%'.format),
                              'Overall Passing Rate': overallschool.map('{:.2f}%'.format)})
school_summary
```

```python
top_schools = school_summary.sort_values(['Overall Passing Rate'], ascending=False)
top_schools.head()
```
